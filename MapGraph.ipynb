```python
import plotly.express as pltx
import plotly.io as pio
import json
import pandas as pd
```

```python
import numpy as np
```

```python
list(pio.renderers)
```

```python
india=json.load(open("/Users/j.sanyam/Downloads/states_india.geojson","r"))
census1=pd.read_csv("/Users/j.sanyam/Downloads/india_census.csv")
pio.renderers.default='chrome'
```

```python
india['features'][0]
```

```python
census1.head()
```

```python
id_map={}
for feature in india['features']:
    feature['id']=feature['properties']['state_code']
    id_map[feature['properties']['st_nm']]=feature['id']
```

```python
id_map
```

```python
census1['id']=census1['State or union territory'].apply(lambda x : id_map[x])
census1['Density']=census1['Density[a]'].apply(lambda x: int(x.split("/")[0].replace(",","")))
census1['DensityScale']=np.log10(census1['Density'])
```

```python
census1.head()
```

```python
indmap=pltx.choropleth(census1,locations='id',geojson=india,color='DensityScale',scope='asia',hover_name='State or union territory',hover_data=['Density','Population'])
indmap.update_geos(fitbounds='locations',visible=False)
indmap.show()
```

```python
indmap1=pltx.choropleth_mapbox(census1,locations='id',geojson=india,color='DensityScale',hover_name='State or union territory',hover_data=['Density','Population'],mapbox_style="carto-positron",center={'lat':24,'lon':78},zoom=3,opacity=0.5)
indmap1.show()
```
